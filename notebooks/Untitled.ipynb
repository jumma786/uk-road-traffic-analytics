```python
import requests
import pandas as pd

base_url = "https://roadtraffic.dft.gov.uk/api/average-annual-daily-flow"

# Get metadata
response = requests.get(base_url, params={"page[size]": 1000})
meta = response.json()

last_page = meta["last_page"]

print(f"Total pages: {last_page}")
print(f"Total records: {meta['total']}")

all_data = []

for page in range(1, last_page + 1):
    response = requests.get(
        base_url,
        params={
            "page[size]": 1000,
            "page[number]": page
        }
    )

    all_data.extend(response.json()["data"])

    print(f"Downloaded {page}/{last_page}")

df = pd.DataFrame(all_data)

print(df.shape)

df.to_csv("uk_traffic_full.csv", index=False)
```
