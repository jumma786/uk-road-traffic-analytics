```python
from pathlib import Path
import pandas as pd

raw_folder = Path("data/raw")

for file in raw_folder.glob("*"):
    try:
        if file.suffix.lower() == ".csv":
            df = pd.read_csv(file, nrows=5)

        elif file.suffix.lower() == ".parquet":
            df = pd.read_parquet(file)

        else:
            continue

        print("=" * 80)
        print(f"File: {file.name}")
        print(f"Columns ({len(df.columns)}):")
        print(list(df.columns))
        print()

    except Exception as e:
        print(f"Error reading {file.name}: {e}")
```

```python
from pathlib import Path
import pandas as pd

raw_folder = Path("data/raw")

summary = []

for file in raw_folder.glob("*.csv"):
    try:
        df = pd.read_csv(file, nrows=5)

        summary.append({
            "file": file.name,
            "columns": len(df.columns),
            "column_names": ", ".join(df.columns)
        })

    except Exception as e:
        summary.append({
            "file": file.name,
            "columns": "ERROR",
            "column_names": str(e)
        })

summary_df = pd.DataFrame(summary)

print(summary_df)
```

```python
import geopandas as gpd
import pandas as pd

# Read shapefile
gdf = gpd.read_file("data/raw/MRDB_2025_published.shp")

# Read traffic data
df = pd.read_csv("data/raw/dft_traffic_counts_aadf_by_direction.csv")

# Get total traffic per count point (sum directions)
traffic_summary = df.groupby('count_point_id')['all_motor_vehicles'].sum().reset_index()

# Join shapefile to traffic data
gdf_merged = gdf.merge(traffic_summary, left_on='CP_Number', right_on='count_point_id', how='left')

# Now each road segment has a traffic volume!
print(gdf_merged[['CP_Number', 'RoadNumber', 'all_motor_vehicles']].head())
```

```python
import geopandas as gpd
import pandas as pd

gdf = gpd.read_file("data/raw/MRDB_2025_published.shp")
df = pd.read_csv("data/raw/dft_traffic_counts_aadf_by_direction.csv")

# Option 1: Latest year only (2025)
df_2025 = df[df['year'] == 2025].copy()

# Sum directions for each count point in 2025
traffic_2025 = df_2025.groupby('count_point_id')['all_motor_vehicles'].sum().reset_index()

# Join
gdf_merged = gdf.merge(traffic_2025, left_on='CP_Number', right_on='count_point_id', how='left')

print(gdf_merged[['CP_Number', 'RoadNumber', 'all_motor_vehicles']].head())
```
